```python
# mapping of the survey answers
age_mapping = { #question name 'age'
    'A1': 'Under 18',
    'A2': '18-24',
    'A3': '25-34',
    'A4': '35-44',
    'A5': '45-54',
    'A6': '55-64',
    'A7': '65 or older',
    'A8': 'no answer'
}
gender_mapping ={ # question name 'gender'
    'A1': 'male',
    'A2': 'female',
    'A3': 'no answer',
    '-oth-': 'other'
}
education_mapping = { # question name 'educationlevel'
    'A1': 'No formal education',
    'A2': 'Secondary school diploma (e.g., intermediate school certificate)',
    'A3': 'Vocational or technical training',
    'A4': 'University entrance qualification (e.g., high school diploma, Abitur)',
    'A5': "Bachelor's degree or equivalent",
    'A6': "Master's degree or equivalent",
    'A7': 'Higher academic qualification',
    'A8': 'None of the above',
    'A10': 'no answer'
}

role_mapping = { # question name 'educationstatus'
    'A2': 'prospective',
    'A3': 'prospective',
    'A4': 'enrolled',
    'A5': 'international',
    'A1': 'other',
    'A6': 'other',
    'A7': 'other',
    'A8': 'no answer'
}

chatbot_mapping = { # question name 'familiar'
    'A1': 'Very familiar – I use chatbots frequently',
    'A2': 'Familiar – I use chatbots occasionally',
    'A3': 'Neutral – I have some experience with chatbots',
    'A4': 'Unfamiliar – I have limited experience with chatbots',
    'A5': 'Not at all familiar – I have never used a chatbot',
    'A6': 'no answer'
}
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker


sns.set_theme(style="whitegrid")

# --- Custom colors ---
unired    = '#AC0634'  # 172, 6, 52
unigrey   = '#CFCFCF'  # 207, 207, 207
uniyellow = '#FBB900'  # 251,185,0

# This list will be used by Seaborn to color bars (alternating).
custom_palette = [unired, unigrey]

# --- Load DataFrame ---
df = pd.read_csv('../../data/question_survey/questions_12_12_24_noemail.csv')

# --- Print total participants (all) ---
n_total = len(df)
print(f"Total participants in the study (all entries): {n_total}")

# --- Filter to only those who completed the study (lastpage == 3) ---
df = df[df['lastpage'] == 3]

# --- Print participants who completed ---
n_completed = len(df)
print(f"Participants who completed the study (lastpage == 3): {n_completed}")

# --- Apply mappings ---
df['age'] = df['age'].map(age_mapping)
df['gender'] = df['gender'].map(gender_mapping)
df['educationlevel'] = df['educationlevel'].map(education_mapping)
df['educationstatus'] = df['educationstatus'].map(role_mapping)
df['familiar'] = df['familiar'].map(chatbot_mapping)

# Rename 'gender[other]' to something clearer
df.rename(columns={'gender[other]': 'gender_other'}, inplace=True)

# --- Print Summaries ---
print("\nAge distribution:")
print(df['age'].value_counts(dropna=False))

print("\nGender distribution:")
print(df['gender'].value_counts(dropna=False))

print("\nEducation level distribution:")
print(df['educationlevel'].value_counts(dropna=False))

print("\nEducation status distribution:")
print(df['educationstatus'].value_counts(dropna=False))

print("\nChatbot familiarity distribution:")
print(df['familiar'].value_counts(dropna=False))

print("\nProgram distribution:")
print(df['program'].value_counts(dropna=False))


# Utility function to create a nicely formatted count plot
def plot_countplot(data, x_column, title, palette=None, fig_width=6, fig_height=4, rotation=45, path=None):
    plt.figure(figsize=(fig_width, fig_height))
    ax = sns.countplot(
        data=data, 
        x=x_column, 
        palette=palette,          # or color=unired if you prefer a single color
        order=data[x_column].value_counts().index.dropna(),
        edgecolor='black'         # outlines on bars for a sharper look
    )
    ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))  # Ensure integer ticks on the y-axis
    plt.title(title)
    plt.xticks(rotation=rotation, ha='right')
    plt.tight_layout()
    
    # # Optionally add labels showing the count on top of each bar
    # for container in ax.containers:
    #     ax.bar_label(container, label_type='edge', padding=0, fontsize=10)

    if path:
        plt.savefig(path, bbox_inches='tight') 
    
    plt.show()

def plot_countplot_horizontal(data, y_column, title, palette=None, fig_width=6, fig_height=4, path=None):
    """
    Creates a horizontal bar plot for better readability if the categories have long names.
    """
    plt.figure(figsize=(fig_width, fig_height))
    ax = sns.countplot(
        data=data,
        y=y_column,
        palette=palette,
        order=data[y_column].value_counts().index.dropna(),
        edgecolor='black'
    )
    ax.xaxis.set_major_locator(mticker.MaxNLocator(integer=True))  # Ensure integer ticks on the x-axis
    plt.title(title)
    #ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))
    plt.tight_layout()

    # # Add labels showing the count on the side of each bar
    # for container in ax.containers:
    #     ax.bar_label(container, label_type='edge', padding=0, fontsize=10)
    if path:
        plt.savefig(path, bbox_inches='tight') 
    
    plt.show()

path = '../../data/figures/question_survey/'
# --- Visualizations ---
plot_countplot(df, 'age', "Age Distribution of Participants (Completed)", 
               palette=custom_palette, fig_width=12, fig_height=6, rotation=45, path=path + 'demographic_age.png')

plot_countplot(df, 'gender', "Gender Distribution of Participants (Completed)", 
               palette=custom_palette, fig_width=12, fig_height=6, rotation=45, path=path + 'demographic_gender.png')

plot_countplot(df, 'educationlevel', "Education Level of Participants (Completed)", 
               palette=custom_palette, fig_width=12, fig_height=6, rotation=45, path=path + 'demographic_education_level.png')

plot_countplot(df, 'educationstatus', "Education Status of Participants (Completed)", 
               palette=custom_palette, fig_width=12, fig_height=6, rotation=45, path=path + 'demographic_education_status.png')

plot_countplot(df, 'familiar', "Chatbot Familiarity of Participants (Completed)", 
               palette=custom_palette, fig_width=12, fig_height=6, rotation=45, path=path + 'demographic_chatbot_familiarity.png')
plot_countplot_horizontal(df, 'program', "Program Distribution of Participants (Completed)", 
                          palette=custom_palette, fig_width=12, fig_height=6, path=path + 'demographic_program_distribution.png')
```
