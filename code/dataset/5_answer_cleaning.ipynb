```python
import pandas as pd
import re

def transform_answers_to_long(
    input_csv: str, 
    output_csv: str, 
    max_links: int = 10
):
    """
    Transforms the 'answers_cleaned.csv' wide format into a long format.
    Each row in the resulting DataFrame represents one question–answer pair.
    
    Parameters
    ----------
    input_csv : str
        Path to the wide-format CSV (e.g. 'answers_06_01_cleaned.csv').
    output_csv : str
        Path to save the transformed long-format CSV.
    max_links : int
        Maximum number of link columns per answer, e.g., a1link[SQ001] ... a1link[SQ010].
        Adjust if you expect more or fewer columns.
    """
    
    # Load the cleaned data
    df = pd.read_csv(input_csv)
    
    # Print rows and column Before long formatting
    print("Before long formatting")
    print(f'Rows: {df.shape[0]}')
    print(f'Columns: {df.shape[1]}\n')

    # Identify columns for participant-level metadata 
    # i.e. everything that is *not* specifically for answers (a1, a2, a3),
    # question_ids (q1id, q2id, q3id), or link columns
    answer_cols_pattern = re.compile(r'^(a[123]|q[123]id|a[123]link\[SQ\d+\])')
    participant_cols = [col for col in df.columns if not answer_cols_pattern.match(col)]
    
    # Build a new DataFrame with one row per question–answer pair
    long_data = []
    
    for _, row in df.iterrows():
        # Convert the single participant row into up to three answer rows
        participant_dict = row[participant_cols].to_dict()  # baseline participant info

        # For each answer slot (1,2,3)
        for i in [1, 2, 3]:
            answer_col = f'a{i}'        # e.g. 'a1'
            qid_col    = f'q{i}id'      # e.g. 'q1id'
            
            # If both the answer and question_id are NaN, skip this slot - should not happen
            if pd.isna(row[answer_col]) and pd.isna(row[qid_col]):
                print(f"Skipping empty answer slot {i} for participant {row['participant_id']}")
                continue

            # Extract the question_id for this answer
            question_id = row[qid_col]
            # cast to int
            question_id = int(question_id)
            
            # Gather link columns for this answer (e.g. a1link[SQ001], a1link[SQ002], etc.)
            # Use three-digit zero-padding to handle 'SQ001' through 'SQ010' or more
            link_cols = [f'a{i}link[SQ{n:03d}]' for n in range(1, max_links + 1)]
            
            # Filter out only columns that actually exist in df
            valid_link_cols = [col for col in link_cols if col in df.columns]
            
            # Extract the link values (dropping NaNs)
            link_values = [row[col] for col in valid_link_cols if not pd.isna(row[col])]
            
            # Build the new row as a dictionary
            answer_dict = {
                **participant_dict,           # all participant-level data
                'question_id': question_id,
                'answer_text': row[answer_col],
            }
            
            # Store all links in a single string (semicolon-separated) 
            answer_dict['answer_links'] = '; '.join(link_values)
            
            long_data.append(answer_dict)
    
    # Convert the list of dictionaries to a DataFrame
    long_df = pd.DataFrame(long_data)
    
    # Reorder columns for clarity
    main_cols = [
        'answer_text', 'answer_links', 'question_id', 'participant_id',  'role'
    ]
    other_cols = [c for c in long_df.columns if c not in main_cols]
    ordered_cols = main_cols + other_cols
    long_df = long_df[ordered_cols]

    # Print rows and column After long formatting
    print("After long formatting")
    print(f'Rows: {long_df.shape[0]}')
    print(f'Columns: {long_df.shape[1]}\n')

    # Add the suffix '_a' to all columns for destinguishing between the question dataset
    long_df.columns = [f'{col}_a' for col in long_df.columns]

    # Save to CSV
    long_df.to_csv(output_csv, index=False, quoting=1)
    print(f"Long-format answers saved to {output_csv}")
    return long_df

# Transform the cleaned data to long format
long_df = transform_answers_to_long(
    input_csv="../../data/answer_survey/answers_06_01_cleaned.csv",
    output_csv="../../data/answer_survey/answers_06_01_long_links.csv"
)

long_df.head()
```

```python
import os
import re
from openai import OpenAI
from tqdm import tqdm
import pandas as pd

def translate_text(text):
    """
    Translates a given text to English using the OpenAI Chat Completion API (GPT-4).
    """
    try:
        prompt = f"Translate the following text to English:\n\"\"\"\n{text}\n\"\"\""
        system_message = {
            "role": "system",
            "content": (
                "You are a translation assistant. Please translate the user's input as accurately as possible,\
                ensuring that the original meaning is preserved. Maintain all punctuation, formatting, and any \
                special characters exactly as they appear in the original text. Do not add any additional information.\
                If the text is already in English, please copy it as is. OUTPUT ONLY THE TRANSLATED TEXT WITHOUT \
                SPECIAL CHARACTERS BEFORE OR AFTER."
            )
        }

        response = client.chat.completions.create(
            messages=[
                system_message,
                {"role": "user", "content": prompt}
            ],
            model="gpt-4o-2024-08-06",
            temperature=0
        )
        return response.choices[0].message.content.strip(), True
    except Exception as e:
        print(f"Error translating text: {e}")
        return text, False  # Return original text if there's an error

def translate_text_in_brackets(text):
    """
    1. Looks for occurrences of [some text](some link).
    2. For each match, translates only the bracketed text.
    3. Replaces the bracketed text with its translation, leaving the link unchanged.
    """

    # This function will replace each match as we go along
    def _replace_bracketed(match):
        bracketed_text = match.group(1)  # the text inside [  ]
        link_part = match.group(2)       # the link inside (  )

        translated_bracketed, success = translate_text(bracketed_text)
        if not success:
            # If there was an error, keep the original bracketed text
            translated_bracketed = bracketed_text

        # Rebuild in the format [translated text](original link)
        return f"[{translated_bracketed}]({link_part})"

    # --- REGEX to find bracket-text plus parentheses-link: [text](link) ---
    pattern_bracket_link = re.compile(r'\[([^]]+)\]\(([^)]+)\)')

    # Now apply the sub function to the entire string
    translated_string = pattern_bracket_link.sub(_replace_bracketed, text)
    return translated_string
```

```python
# Translating the answers into English

# Load the environment variables
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

# Load the CSV file
file_path = '../../data/answer_survey/answers_06_01_long_links.csv'
answer_data = pd.read_csv(file_path)

# 1) Translate bracketed text in answer_links_a ------------------------
translated_links_col = []
translation_links_success = []

for _, row in tqdm(answer_data.iterrows(), total=answer_data.shape[0]):
    original_links = row['answer_links_a']

    # If the cell is NaN or empty, skip
    if not isinstance(original_links, str) or not original_links.strip():
        translated_links_col.append(original_links)
        translation_links_success.append(False)
        continue

    try:
        # Translate only the bracketed text
        translated_links = translate_text_in_brackets(original_links)
        translated_links_col.append(translated_links)
        translation_links_success.append(True)
        
        print("Original links:\n", original_links)
        print("*" * 50)
        print("Translated links:\n", translated_links)
        print("-" * 50)

    except Exception as e:
        print(f"Error translating links: {e}")
        # Fallback: just keep the original if there's an error
        translated_links_col.append(original_links)
        translation_links_success.append(False)

# 2) Translate answer_text_a ----------------
translated_answers = []
translation_success = []

for _, row in tqdm(answer_data.iterrows(), total=answer_data.shape[0]):
    original_text = row['answer_text_a']
    translated_text, success = translate_text(original_text)

    translated_answers.append(translated_text)
    translation_success.append(success)

    print("Original:\n", original_text)
    print("*" * 50)
    print("Translated:\n", translated_text)
    print("-" * 50)

answer_data['english_answer_text_a'] = translated_answers
#answer_data['translation_done_a'] = translation_success


answer_data['english_answer_links_a'] = translated_links_col
#answer_data['translation_done_links_a'] = translation_links_success

# Save to a new CSV file (update path/filename as you like)
output_path_translated = '../../data/answer_survey/answer_dataset.csv'
answer_data.to_csv(output_path_translated, index=False, quoting=1)

# Finally, display a sample of the updated DataFrame
answer_data.head()
```
